```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Data Files

```python
# Load the data
file_path1 = '07-08 to 07-14 PHX RAID Reviews.csv'
data_phx = pd.read_csv(file_path1)

file_path2 = '07-08 to 07-14 ELP RAID Reviews.csv'
data_elp = pd.read_csv(file_path2)

file_path3 = '3Plus Devices Data 07-08 to 07-14.csv'
three_plus_df = pd.read_csv(file_path3)
```

```python

# Split three_plus_df
three_plus_elp = three_plus_df[three_plus_df['Site'] == 'ELP']
three_plus_phx = three_plus_df[three_plus_df['Site'] == 'PHX']
```

```python

# Group by "Worked By" and "Action Taken" and count the occurrences
grouped_data_phx = data_phx.groupby(['Worked By', 'Action Taken']).size().unstack(fill_value=0)

# Group by "Worked By" and "Action Taken" and count the occurrences
grouped_data_elp = data_elp.groupby(['Worked By', 'Action Taken']).size().unstack(fill_value=0)

# Group by "Worked By" and "Action Taken" and count the occurrences
grouped_three_plus_phx = three_plus_phx.groupby(['Worked By', 'Action Taken']).size().unstack(fill_value=0)

# Group by "Worked By" and "Action Taken" and count the occurrences
grouped_three_plus_elp = three_plus_elp.groupby(['Worked By', 'Action Taken']).size().unstack(fill_value=0)
```

### Get Dates For Previous Week (Monday - Sunday)

```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Step 1: Calculate the previous week's Monday and Sunday dates
today = datetime.today()
last_sunday = today - timedelta(days=today.weekday() + 1)
last_monday = last_sunday - timedelta(days=6)

# Step 2: Format the dates to the desired string format (e.g., "June 10th")
def format_date(date):
    return date.strftime("%B %d").replace(' 0', ' ')

monday_str = format_date(last_monday)
sunday_str = format_date(last_sunday)
```

### Create Graphs

```python

# Create the stacked bar chart with approved and rejected quantities stacked
ax = grouped_data_phx.plot(kind='bar', stacked=True, color=['green', 'yellow', 'red'])

# Title and labels
plt.title(f"Phoenix Yellow Path 2.0 Reviews {monday_str} through {sunday_str}")
plt.xlabel("Worked By")
plt.ylabel("Number of Actions")
plt.xticks(rotation=90)  # Rotate the 'Worked By' labels vertically

# Show the plot
plt.show()
```

```python

# Create the stacked bar chart with approved and rejected quantities stacked
ax = grouped_data_elp.plot(kind='bar', stacked=True, color=['green', 'yellow', 'red'])

# Title and labels
plt.title(f"El Paso Yellow Path 2.0 Reviews {monday_str} through {sunday_str}")
plt.xlabel("Worked By")
plt.ylabel("Number of Actions")
plt.xticks(rotation=90)  # Rotate the 'Worked By' labels vertically

# Show the plot
plt.show()
```

```python
# Create the stacked bar chart with approved and rejected quantities stacked
ax = grouped_three_plus_phx.plot(kind='bar', stacked=True, color=['green', 'yellow', 'red'])

# Title and labels
plt.title(f"Phoenix 3+ Worker Tool Reviews {monday_str} through {sunday_str}")
plt.xlabel("Worked By")
plt.ylabel("Number of Actions")
plt.xticks(rotation=90)  # Rotate the 'Worked By' labels vertically

# Show the plot
plt.show()
```

```python
# Create the stacked bar chart with approved and rejected quantities stacked
ax = grouped_three_plus_elp.plot(kind='bar', stacked=True, color=['green', 'yellow', 'red'])

# Title and labels
plt.title(f"El Paso 3+ Worker Tool Reviews {monday_str} through {sunday_str}")
plt.xlabel("Worked By")
plt.ylabel("Number of Actions")
plt.xticks(rotation=90)  # Rotate the 'Worked By' labels vertically

# Show the plot
plt.show()
```

```python
# Group by "Worked By" and "Action Taken" and count the occurrences
grouped_data_phx = data_phx.groupby(['Worked By', 'Action Taken']).size().unstack(fill_value=0)

# Calculate the percentage of 'Rejected' per person
grouped_data_phx['Total'] = grouped_data_phx.sum(axis=1)
grouped_data_phx['Rejected_Percentage'] = (grouped_data_phx['Hold'] / grouped_data_phx['Total']) * 100

# Plotting the percentage of 'Rejected' per person
ax = grouped_data_phx['Rejected_Percentage'].plot(kind='bar', color='teal')

# Title and labels
plt.title(f"Yellow Path 2.0 Percentage 'Held' per Person ({monday_str} to {sunday_str})")
plt.xlabel("Worked By")
plt.ylabel("Percentage of Hold Actions")
plt.xticks(rotation=90)  # Rotate the 'Worked By' labels vertically

# Show the plot
plt.show()
```

```python
# Group by "Worked By" and "Action Taken" and count the occurrences
grouped_data_elp = data_elp.groupby(['Worked By', 'Action Taken']).size().unstack(fill_value=0)

# Calculate the percentage of 'Rejected' per person
grouped_data_elp['Total'] = grouped_data_elp.sum(axis=1)
grouped_data_elp['Rejected_Percentage'] = (grouped_data_elp['Hold'] / grouped_data_elp['Total']) * 100

# Plotting the percentage of 'Rejected' per person
ax = grouped_data_elp['Rejected_Percentage'].plot(kind='bar', color='teal')

# Title and labels
plt.title(f"Yellow Path 2.0 Percentage 'Held' per Person ({monday_str} to {sunday_str})")
plt.xlabel("Worked By")
plt.ylabel("Percentage of Hold Actions")
plt.xticks(rotation=90)  # Rotate the 'Worked By' labels vertically

# Show the plot
plt.show()
```
